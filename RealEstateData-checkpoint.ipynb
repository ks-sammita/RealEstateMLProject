Feature engineering on mdf1

```python
from sklearn.pipeline import Pipeline,FeatureUnion
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import FunctionTransformer
numerical_features = ['ZIP OR POSTAL CODE', 'BEDS', 'BATHS', 'SQUARE FEET', 'LOT SIZE', 'YEAR BUILT', 'DAYS ON MARKET','LOT_SQ','sq_zp','ls_zp']
categorical_features = ['PROPERTY TYPE', 'CITY']
lot_size_sqft_interaction = mdf1['LOT SIZE'] * mdf1['SQUARE FEET']
sqft_zip_ratio = mdf1['SQUARE FEET'] / mdf1['ZIP OR POSTAL CODE']
lot_size_zip_ratio = mdf1['LOT SIZE'] / mdf1['ZIP OR POSTAL CODE']
mdf1.loc[:,'LOT_SQ']=lot_size_sqft_interaction
mdf1.loc[:,'sq_zp']=sqft_zip_ratio
mdf1.loc[:,'ls_zp']=lot_size_zip_ratio
print(mdf1.head())
preprocessor = ColumnTransformer(
    transformers=[
#         ('num', StandardScaler(), numerical_features),  
        ('cat', OrdinalEncoder(), categorical_features)  # Encode categorical features
    ],
    remainder='passthrough' 
)




# Pipeline with preprocessing and additional feature engineering
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
    ])

# Fit and transform the data
transformed_data = pipeline.fit_transform(mdf1)
print(transformed_data)
# Convert back to DataFrame for verification
mdf2 = pd.DataFrame(transformed_data, columns=numerical_features + categorical_features+['PRICE'])



print(mdf2.head())
```
